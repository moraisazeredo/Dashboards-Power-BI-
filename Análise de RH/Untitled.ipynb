```python
import pandas as pd
import numpy as np

def gerar_datas_aleatorias(inicio, fim, quantidade):
    inicio_ts = pd.Timestamp(inicio).value // 10**9  # Converte para timestamp Unix
    fim_ts = pd.Timestamp(fim).value // 10**9
    timestamps_aleatorios = np.random.randint(inicio_ts, fim_ts, quantidade)
    datas_aleatorias = pd.to_datetime(timestamps_aleatorios, unit='s') 
    return datas_aleatorias

def criar_dataframe_com_datas(inicio, fim, quantidade):
    datas_aleatorias = gerar_datas_aleatorias(inicio, fim, quantidade)
    df = pd.DataFrame({'Data Venda': datas_aleatorias})
    return df

# Exemplo de uso
inicio = '2014-12-01'
fim = '2020-06-15'
quantidade = 100

# Criar e imprimir o DataFrame
df_datas = criar_dataframe_com_datas(inicio, fim, quantidade)
```

```python
ids = [4, 8, 9, 41, 42]
valor_venda = np.random.randint(50, 1000, size=len(df_datas))

df_datas['Valor Venda'] = np.random.choice(valor_venda, size=len(df_datas))
df_datas['ID'] = np.random.choice(ids, size=len(df_datas))
```
